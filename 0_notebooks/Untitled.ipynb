## Engenharia de Features

```python
def convert_to_metros(value):
    return value/10.764

ls_colunas_para_transformacao = ['1stflrsf','2ndflrsf']

for coluna in ls_colunas_para_transformacao:
    data_final[coluna] = data_final[coluna].apply(convert_to_metros)
```

```python
data_final['idade_da_casa'] = 2021 - data_final.yearbuilt
```

```python
data_final.drop(['yearbuilt'],axis=1,inplace=True)
```

# Modelagem

```python
## importacoes

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
x,y = data_final.drop(['saleprice'],axis=1), data_final.saleprice
```

```python
y_pred_media = [y.mean() for _ in range(len(y))]
print('MSE  : ',mean_squared_error(y,y_pred_media))
print('RMSE : ',mean_squared_error(y,y_pred_media)**0.5)
```

```python
## train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
## MODELO DE REGRESSAO LINEAR MULTIPLA

lr = LinearRegression(n_jobs=-1)
lr.fit(x_train,y_train)
```

```python
## avaliando metricas para teste

y_pred_teste = lr.predict(x_test)

print('MSE  : ',mean_squared_error(y_test,y_pred_teste))
print('RMSE : ',mean_squared_error(y_test,y_pred_teste)**0.5)
```

```python
## avaliando metricas para treino

y_pred_treino = lr.predict(x_train)

print('MSE  : ',mean_squared_error(y_train,y_pred_treino))
print('RMSE : ',mean_squared_error(y_train,y_pred_treino)**0.5)
```

```python
## r2

r2_score(y_test,y_pred_teste)
```

```python
dc_data = {}


for i in range(10,90,5):
    test_size = i/100
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=test_size,random_state=42)
    
    lr = LinearRegression(n_jobs=-1)
    lr.fit(x_train,y_train)
    
    dc_data[test_size] = [mean_squared_error(y_test,lr.predict(x_test)),
                          mean_squared_error(y_train,lr.predict(x_train)),
                          r2_score(y_test,lr.predict(x_test))]
```

```python
f, ax = plt.subplots(2,1,sharex=True,figsize=(16,9))


ax[0].plot(dc_data.keys(),[x[0] for x in dc_data.values()],color='tab:orange',label='mse de teste')
ax[0].plot(dc_data.keys(),[x[1] for x in dc_data.values()],color='tab:blue',label='mse de treino')
ax[0].legend()

ax[1].plot(dc_data.keys(),[x[2] for x in dc_data.values()],color='black',label='r2 de teste')
ax[1].set_xticks(list(dc_data.keys()))
ax[1].legend()

plt.legend()
```

```python
def avaliacao_modelos(modelo1,modelo2,x,y):
    
    ## quebrando o nosso dataset em treino e teste
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)
    
    ### TREINANDO MODELO 1
    modelo1.fit(x_train,y_train)
    
    ### TREINANDO MODELO 2
    modelo2.fit(x_train,y_train)
    
    ### AVALIAR OS MODELOS
    
    mse = []
    r2  = []
    
    mse.append(mean_squared_error(y_test,modelo1.predict(x_test)))
    mse.append(mean_squared_error(y_test,modelo2.predict(x_test)))
    
    r2.append(r2_score(y_test,modelo1.predict(x_test)))
    r2.append(r2_score(y_test,modelo2.predict(x_test)))
    
    df_resultados = pd.DataFrame([mse,r2],columns=['modelo1','modelo2'],index=['mse','r2'])
    
    return df_resultados, modelo1, modelo2
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor(min_samples_leaf=10,random_state=991151575)
```

```python
df_resultados, modelo1, modelo2 = avaliacao_modelos(lr,dt,x,y)
```
