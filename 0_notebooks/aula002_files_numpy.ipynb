```python
## outras funções importantes

arr = np.array(np.random.uniform(size=32)).reshape(16,2)
arr
```

```python
arr = arr.flatten()
arr[5] = 99
IQR = np.quantile(arr,0.75) - np.quantile(arr,0.25)
q1 = np.quantile(arr,0.25)

limite_inferior = q1 - 1.5 * IQR
limite_superior = np.quantile(arr,0.75) + 1.5 * IQR

arr[ (arr < limite_superior) & (arr > limite_inferior) ]
```
